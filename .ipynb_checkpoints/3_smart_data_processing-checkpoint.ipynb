# LABELS

Our labels will be one random lactate sampled from each admission with 2 or more lactates, 2nd lactate or later

To get to to eligible values, we will...
1. Use only quantitative data
2. We will combine all sources of quantitative data into a single columns based on UOM
3. We will keep only the largest of these columns
4. We will keep only admissions with 2 or more lactate values

Full label pipeline...

```python
import features
import transformers
from constants import column_names,ALL
from sklearn.pipeline import Pipeline
from constants import variable_type
```

```python
#will ensure that there is at least 1 lactate before the sampled lactate
class keep_one_before_sample(features.preserve_datetime_sample):
    
    def transform(self, df, **transform_params):
        df = df.groupby(level=column_names.ID).apply(lambda grp: grp.iloc[1:]).reset_index(0,drop=True)
        return super(keep_one_before_sample,self).transform(df)
```

```python
next_lactate_sampler = keep_one_before_sample(levels=[column_names.ID],random_state=random_state)

labels = features.Feature(name=next_lactate_sampler.name, 
                             data_needs=[(data_dict.components.LACTATE,ALL)],
                             col_filter=transformers.filter_to_component([data_dict.components.LACTATE]),
                             aggregator=next_lactate_sampler)
label_cleaner = Pipeline([
        ('quantitative only',transformers.filter_var_type([variable_type.QUANTITATIVE])),
        ('combine_like_columns',transformers.combine_like_cols()),
        ('max_col',transformers.max_col_only()),
        ('two_or_more_lactate',transformers.more_than_n_component(1,data_dict.components.LACTATE)),
        ('dropna',transformers.dropna()),
    ]) 

next_lactate_labelizer = features.Featurizer(index_levels=[column_names.ID], 
                                                   loader=loader,features=[labels],
                                                   pre_cleaners=label_cleaner)
next_lactate_labelizer
```

# Features

```python
import numpy as np
from sklearn.pipeline import Pipeline
import load_and_segment
```

```python
def add_simple_feature(featurizer,component,agg_func,fillna_method):
    featurizer.add_feature(
        name=agg_func.func_name.upper(),
        data_needs=[(component,ALL)],
        col_filter=transformers.filter_to_component([component]),
        agg_func=agg_func,
        fillna_method=fillna_method,
    )
    return featurizer

"""
HOW TO CLEAN DATA BEFORE FEATURES
"""
pre_cleaners = Pipeline([
            ('drop_small_columns',transformers.remove_small_columns(threshold=50)),
            ('drop_low_id_count',transformers.record_threshold(threshold=20)),
            ('quantitative only',transformers.filter_var_type([variable_type.QUANTITATIVE])),
            ('combine_like_columns',transformers.combine_like_cols()),
        ])

"""
FEATURES
"""
bp_featurizer = load_and_segment.SegmentFeaturizer(loader,segmenter=None,pre_cleaners=pre_cleaners)

fill_mean = transformers.fill_mean
fill_zero = transformers.fill_zero
def count(grp): return grp.count()

component = data_dict.components.BLOOD_PRESSURE_SYSTOLIC
add_simple_feature(bp_featurizer,component,np.mean,fill_mean())
add_simple_feature(bp_featurizer,component,np.std,fill_zero())
add_simple_feature(bp_featurizer,component,count,fill_zero())
add_simple_feature(bp_featurizer,component,features.last,fill_mean())

component = data_dict.components.BLOOD_PRESSURE_DIASTOLIC
add_simple_feature(bp_featurizer,component,np.mean,fill_mean())
add_simple_feature(bp_featurizer,component,np.std,fill_zero())
add_simple_feature(bp_featurizer,component,count,fill_zero())
add_simple_feature(bp_featurizer,component,features.last,fill_mean())

component = data_dict.components.BLOOD_PRESSURE_MEAN
add_simple_feature(bp_featurizer,component,np.mean,fill_mean())
add_simple_feature(bp_featurizer,component,np.std,fill_zero())
add_simple_feature(bp_featurizer,component,count,fill_zero())
add_simple_feature(bp_featurizer,component,features.last,fill_mean())
```

# Models!

```python
from sklearn.metrics import mean_squared_error
import pandas as pd

def try_model(train_ft,train_lbl,test_ft,test_lbl,prep_pipeline,model):
    # apply final cleaning to our data
    X_train = prep_pipeline.fit_transform(train_ft.copy())
    y_train = train_lbl.reset_index(drop=True).iloc[:,0]
    X_test = prep_pipeline.fit_transform(test_ft.copy())
    y_test = test_lbl.reset_index(drop=True).iloc[:,0]
    
    model = model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    
    rmse = pd.np.sqrt(mean_squared_error(y_test,y_pred))

    results = pd.DataFrame(zip(y_test,y_pred), columns=['actual','predicted'])
    
    display(results.head())
    display(results.describe())
    print("Training set score: %f" % model.score(X_train, y_train))
    print("Test set score: %f" % model.score(X_test, y_test))
    
    print "RMSE:",rmse
    
    return model,results,rmse
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

prep_pipeline = Pipeline([
        ('flatten_index',transformers.flatten_index(axis=0)),
        ('flatten_columns',transformers.flatten_index(axis=1)),
        ('scaler',StandardScaler())
    ])

lin_reg = LinearRegression()


model,results,rmse = try_model(X_train,y_train,X_test,y_test,prep_pipeline,lin_reg)
```

```python
pd.Series(model.coef_, index=X_train.columns).sort_values().to_frame()
```

# Optimizing segmetation

```python
import utils
import icu_data_defs
import load_and_segment
import pandas as pd
```

```python
#load data
df_lactate = utils.open_df('data/mimic_data','cleaned/lactate')
df_heart_rate = utils.open_df('data/mimic_data','cleaned/heart rate')
```

```python
df_both = df_lactate.head(500).join(df_heart_rate.head(500), how='outer')
df_both.head()
```

```python
periodic_seg = load_and_segment.periodic(n_hrs=5)
df_segments = periodic_seg._segmenter__segment(df_both).reset_index(drop=False)
df_segments.head()
```

```python
df_dt = df_both.drop(df_both.columns, axis=1).reset_index(drop=False)
df_dt.columns = df_dt.columns.get_level_values('component').tolist()
df_dt.head()
```

```python
df_combined = df_segments.merge(df_dt,on='id')
after_start = pd.isnull(df_combined['start_dt']) | (df_combined['datetime'] >= df_combined['start_dt'])
before_end = pd.isnull(df_combined['end_dt']) | (df_combined['datetime'] < df_combined['end_dt'])
in_seg = after_start & before_end
df_combined['in_seg'] = in_seg                 
```

```python
def clean_seg(seg):
    has_values = seg['in_seg'].any()
    
    if has_values: return seg[seg['in_seg']]
    
    start_dt = seg['start_dt'].iloc[0]
    end_dt = seg['end_dt'].iloc[0]

    in_seg_dt = start_dt
    if pd.isnull(in_seg_dt):
        in_seg_dt = end_dt - pd.Timedelta(value=1,unit='s')
    
    seg = seg.iloc[[0]]
    seg.datetime = in_seg_dt
    
    return seg.iloc[[0]]
```

```python
df_combined.groupby(['id','seg_id']).apply(clean_seg).reset_index(level=2,drop=True)
```
